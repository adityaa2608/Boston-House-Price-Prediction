```python
from matplotlib.animation import FuncAnimation 
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler 
from sklearn import metrics
from math import sqrt
from sklearn.linear_model import LinearRegression 
```

```python
## Dividing the dataset into train and test data
from sklearn.model_selection import train_test_split
train_X = boston_2.drop(['MEDV'],axis=1)
train_y = boston_2['MEDV'] 
```

```python
## train test split
X_train,X_test,y_train,y_test = train_test_split(train_X,train_y,test_size=0.3,random_state=400)
```

```python
#Representing LinearRegression as lr(Creating LinearRegression Object)
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
## Predicting the response for the test data
y_pred = lm.predict(X_test)
```

```python
boston_3 = pd.DataFrame(data = [y_pred,y_test]) 
```

```python
boston_3.T
```

```python
## Coefficient of determination of the regressor model
lm.score(X_test,y_test)
```

```python
## R squared value for training data
print('R2 Value:',metrics.r2_score(y_train, lm.predict(X_train)))
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = sqrt(mse)
print('Mean_Squared_Error:' ,mse)
print('Root_Mean_Squared_Error:' ,rmse)
```

Random Forests vs Linear regression Random forests often prove themselves to be both reliable and effective, and are now part of any modern predictive modelers toolkit. Random forests can frequently outperform linear regression. Of course, better performance for out of sample data (test data) is the key metric. There are situations however where a linear regression can also do better than a random forest. These include:

when the underlying function is actually linear when there are a very large number of features

Random Forests thrive at modeling linear combinations of a large number of features. This perhaps might be true here for the Boston House Price Dataset.Random Forests may sometimes require less pre-modelling. You might not have to scale your data, or apply monotonic transformations (log etc). You often dont even have to fiddle around by removing outliers or agonize on what outliers to leave in.

You can incorporate categorical features without much fuss, and automatic partition of the data are easily tolerated. A lot of the time, you will simply register higher accuracy with RF. It is not unusual to spend an afternoon re-engineering your dataset, and still get a mid 70s% r-squared. With typically less fuss using a from sklearn.ensemble import RandomForestRegressor you may score a neat low 80s% r-squared.

# Learning Curves

The following code cell produces four graphs for a decision tree model with different maximum depths. Each graph visualizes the learning curves of the model for both training and testing as the size of the training set is increased. Note that the shaded region of a learning curve denotes the uncertainty of that curve (measured as the standard deviation). The model is scored on both the training and testing sets using R2, the coefficient of determination.

Linear Regression Model 

```python
#Producing learning curve for varying training set sizes and maximum depths 
from sklearn.model_selection import learning_curve
from matplotlib import pyplot 

train_sizes, train_scores, cv_scores = learning_curve(lm,X_train,y_train,train_sizes = np.linspace(.1,1.0,10))

train_scores_mean = np.mean(train_scores,axis=1)
cv_scores_mean = np.mean(cv_scores, axis=1)

pyplot.ylim((-1,1.0))

pyplot.plot(train_sizes, train_scores_mean, 'ro-', label="Train score")
pyplot.plot(train_sizes, cv_scores_mean, 'go-', label="CV score")
pyplot.legend()
```
